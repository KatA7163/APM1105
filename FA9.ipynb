```
import pandas as pd
from scipy.stats import skew, kurtosis

data = {
    'rank':pd.Series(['Prof', 'Prof', 'Prof', 'Prof', 'Prof', 'AssocProf', 'Prof', 'Prof', 'Prof', 'AsstProf', 'Prof', 'Prof', 'Prof', 'AsstProf', 'Prof', 'Prof', 'Prof', 'AsstProf', 'AsstProf', 'Prof', 'Prof', 'AsstProf', 'Prof', 'Prof', 'AsstProf', 'Prof', 'Prof', 'Prof', 'Prof', 'AsstProf', 'Prof', 'AssocProf', 'Prof', 'Prof', 'Prof', 'AssocProf', 'Prof', 'Prof', 'Prof', 'AsstProf', 'Prof', 'Prof', 'AssocProf', 'AsstProf', 'AsstProf', 'Prof', 'Prof', 'AsstProf', 'AssocProf', 'Prof', 'Prof', 'AsstProf', 'Prof', 'Prof', 'AsstProf', 'AssocProf', 'AsstProf', 'AssocProf', 'AsstProf', 'Prof', 'AssocProf', 'AsstProf', 'AssocProf', 'AssocProf', 'Prof', 'AssocProf', 'Prof', 'AsstProf', 'Prof', 'AsstProf', 'Prof', 'AsstProf', 'AssocProf', 'Prof', 'Prof', 'AssocProf', 'Prof', 'Prof', 'Prof']),
    'salary':pd.Series([186960, 93000, 110515, 131205, 104800, 122400, 81285, 126300, 94350, 57800, 128250, 134778, 88000, 162200, 153750, 150480, 75044, 92000, 107300, 150500, 92000, 103106, 73000, 85000, 91100, 99418, 148750, 155865, 91300, 123683, 107008, 155750, 103275, 120000, 119800, 126933, 146856, 102000, 91000, 129000, 137000, 74830, 80225, 77000, 151768, 140096, 74692, 103613, 111512, 122960, 97032, 127512, 105000, 73500, 62884, 72500, 77500, 72500, 71065, 161101, 105450, 104542, 124312, 109954, 109646]),
}

df = pd.DataFrame(data)

grouped = df.groupby('rank')['salary']

descriptive_stats = grouped.describe()

skewness = grouped.skew()
kurtosis = df.groupby('rank').apply(pd.DataFrame.kurt)
emean = grouped.sem()

# Rename columns
descriptive_stats = descriptive_stats.rename(columns={
    'count': 'Valid',
    'mean': 'Mean',
    '50%': 'Median',
    'std': 'Std. Deviation',
    'min': 'Minimum',
    'max': 'Maximum',
    '25%': '25th percentile',
    '75%': '75th percentile'
})

descriptive_stats['Skewness'] = skewness
descriptive_stats['Kurtosis'] = kurtosis
descriptive_stats['Std. Error of Mean'] = emean
descriptive_stats['Missing'] = descriptive_stats['Valid'] - descriptive_stats['Valid']
descriptive_stats['Coefficient of variation'] = descriptive_stats['Std. Deviation'] / descriptive_stats['Mean']
descriptive_stats['10th percentile'] = grouped.quantile(0.1)
descriptive_stats['20th percentile'] = grouped.quantile(0.2)
descriptive_stats['30th percentile'] = grouped.quantile(0.3)
descriptive_stats['40th percentile'] = grouped.quantile(0.4)
descriptive_stats['50th percentile'] = grouped.quantile(0.5)
descriptive_stats['60th percentile'] = grouped.quantile(0.6)
descriptive_stats['70th percentile'] = grouped.quantile(0.7)
descriptive_stats['80th percentile'] = grouped.quantile(0.8)
descriptive_stats['90th percentile'] = grouped.quantile(0.9)

#Seperate else some values do not appear
print(descriptive_stats[['Valid', 'Missing', 'Median', 'Mean', 'Std. Error of Mean', 'Std. Deviation', 'Coefficient of variation', 
                         'Skewness', 'Kurtosis', 'Minimum', 'Maximum', '25th percentile', '50th percentile'
                         ]])
print(descriptive_stats[['75th percentile','10th percentile', '20th percentile', '30th percentile', 
                         '40th percentile', '50th percentile', '60th percentile', '70th percentile', '80th percentile', '90th percentile']])
```
